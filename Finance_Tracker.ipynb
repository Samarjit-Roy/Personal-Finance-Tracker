```python
from tabulate import tabulate
import os
import json
from datetime import datetime

#-------Login detail------
print("-----Login Page----")
def login(users):
#While is use to run the code until name or password are type right...
    while True:
        username=input("Enter Username: ")
        password=input("Enter Password: ")
        for u in users:
            if username==u[0]:
                if password==u[1]:
                    return username
        print("Try It Again..")

#Define Usename And Password Below...
users=[['Samarjit','roy']]

username=login(users)
print(username,"Has logged in")

c="------Welcome Sir-------"
d=c.center(50)
print(d)
#------Login Detail End-----


# The file where we'll save our data
FILENAME = "my_finances.json"
transactions = []

# --- File Handling ---
def save_data():
    """Saves the transactions list to a file."""
    with open(FILENAME, "w") as f:
        json.dump(transactions, f, indent=4)
    print("💾 Data saved successfully!")

def load_data():
    """Loads transactions from a file if it exists."""
    global transactions
    try:
        with open(FILENAME, "r") as f:
            transactions = json.load(f)
        print("📂 Data loaded successfully!")
    except FileNotFoundError:
        print("No data file found. Starting fresh.")
    except json.JSONDecodeError:
        print("Could not read data file. Starting fresh.")
        transactions = []

#------code Start-------
def add_transaction():
    date_str = input("Enter date (YYYY-MM-DD): ")
    category = input("Enter category (Income/Expense): ").capitalize()
    description = input("Enter description: ")
    amount = float(input("Enter amount: "))

    transaction = {
        "date": date_str,
        "category": category,
        "description": description,
        "amount": amount
    }
    transactions.append(transaction)
    print("✅ Transaction added!")

def view_transactions():
  if not transactions:
        print("⚠ No transactions recorded.")
        return
  print("\n--- Transaction List ---")
  # Prepare data for tabulate with index starting from 1
  table_data = [[i + 1, t['date'], t['category'], t['description'], f"\u20B9{t['amount']:.2f}"] for i, t in enumerate(transactions)]
  headers = ["Srno", "Date", "Category", "Description", "Amount"]
  print(tabulate(table_data, headers=headers, tablefmt='fancy_grid'))

def search():
  while True:
    print('''\n======= Search Transactions ======
1. Search by Category
2. Search by description
3. Search by date
4. Search by amount
5. Go back\n''')
    choice = input("Enter your choice (1-5): ")
    if choice == '1':
      search_category()
    elif choice == '2':
      search_description()
    elif choice == '3':
      search_date()
    elif choice == '4':
      search_amount()
    elif choice == '5':
      # Removed the extra main() call here
      break # Exit the search loop
    else:
      print("\n❌ Invalid choice. Please select an option from 1 to 5.")

def search_category():
    keyword = input("Enter keyword to search (Income/Expense): ").upper()
    results = [t for t in transactions if keyword in t['category'].upper()]
    if results:
        print("\n--- Search Results ---")
        for t in results:
            table_data = [[i + 1, t['date'], t['category'], t['description'], f"\u20B9{t['amount']:.2f}"] for i, t in enumerate(results)]
            headers = ["Srno", "Date", "Category", "Description", "Amount"]
            print(tabulate(table_data, headers=headers, tablefmt='fancy_grid'))
    else:
        print("⚠ No matching transactions found.")

def search_description():
    keyword = input("Enter keyword to search: ").lower()
    results = [t for t in transactions if keyword in t['description'].lower()]
    if results:
        print("\n--- Search Results ---")
        for t in results:
            table_data = [[i + 1, t['date'], t['category'], t['description'], f"\u20B9{t['amount']:.2f}"] for i, t in enumerate(results)]
            headers = ["Srno", "Date", "Category", "Description", "Amount"]
            print(tabulate(table_data, headers=headers, tablefmt='fancy_grid'))
    else:
        print("⚠ No matching transactions found.")

def search_date():
    keyword = input("Enter date (YYYY-MM-DD): ")
    results = [t for t in transactions if t['date'] == keyword]
    if results:
      print("\n--- Search Results ---")
      for t in results:
        table_data = [[i + 1, t['date'], t['category'], t['description'], f"\u20B9{t['amount']:.2f}"] for i, t in enumerate(results)]
        headers = ["Srno", "Date", "Category", "Description", "Amount"]
        print(tabulate(table_data, headers=headers, tablefmt='fancy_grid'))
    else:
        print("⚠ No matching transactions found.")

def search_amount():
    keyword = int(input("Enter amount: "))
    results = [t for t in transactions if t['amount'] == keyword]
    print("\n--- Search Results ---")
    if results:
      print("\n--- Search Results ---")
      for t in results:
        table_data = [[i + 1, t['date'], t['category'], t['description'], f"\u20B9{t['amount']:.2f}"] for i, t in enumerate(results)]
        headers = ["Srno", "Date", "Category", "Description", "Amount"]
        print(tabulate(table_data, headers=headers, tablefmt='fancy_grid'))
    else:
        print("⚠ No matching transactions found.")

def finance_summary():
  if not transactions:
    print("\nNo transactions yet.")
  else:
    balance = 0
    print("\n--- All Transactions ---")
    for t in sorted(transactions, key=lambda x: x['date']):
      if t['category'] == 'Income':
        balance += t['amount']
        print(f"✅ {t['date']}: +₹{t['amount']:8.2f} ({t['description']})")
      elif t['category'] == 'Expense':
        balance -= t['amount']
        print(f"🔻 {t['date']}: -₹{t['amount']:8.2f} ({t['description']})")
      print(f"------------------------\nBalance: ₹{balance:.2f}")

# --- 3. Bonus Feature ---

def display_monthly_chart(transactions):
    """Displays an ASCII bar chart of spending per month."""
    monthly_expenses = {}
    for t in transactions:
        if t['category'] == 'Expense':
            month = t['date'][:7]  # Extract YYYY-MM
            monthly_expenses[month] = monthly_expenses.get(month, 0) + t['amount']

    if not monthly_expenses:
        print("\nNo expenses recorded to display in a chart.")
        return

    print("\n--- Monthly Spending Chart ---")
    # Sort months for chronological order
    for month in sorted(monthly_expenses.keys()):
        total = monthly_expenses[month]
        # Scale the bar: one '█' for every ₹500 spent
        bar = '█' * int(total / 500)
        print(f"{month} | {bar} ₹{total:.2f}")

def delete_data():
    """Deletes a transaction by its serial number."""
    if not transactions:
        print("\n⚠ No transactions to delete.")
        return

    # First, show the user the list with Srno to choose from
    view_transactions()

    try:
        # Ask the user which transaction to delete
        sr_no_to_delete = int(input("\nEnter the Srno of the transaction to delete: "))

        # Convert the 1-based Srno to a 0-based list index
        index_to_delete = sr_no_to_delete - 1

        # Check if the entered number is a valid index
        if 0 <= index_to_delete < len(transactions):
            # Show the user what they are about to delete for confirmation
            item = transactions[index_to_delete]
            print(f"\nYou are about to delete: [{item['date']}, {item['description']}, ₹{item['amount']:.2f}]")

            confirm = input("Are you sure you want to delete this? (y/n): ").lower()

            if confirm == 'y':
                # Use pop() to remove the item at the specified index
                transactions.pop(index_to_delete)
                print("✅ Transaction deleted successfully!")
            else:
                print("Deletion cancelled.")
        else:
            # This runs if the user enters a number that is not in the list
            print("❌ Invalid Srno. No transaction with that number exists.")

    except ValueError:
        # This runs if the user enters text instead of a number
        print("❌ Invalid input. Please enter a number.")


def main():
    load_data()
    while True:
        print("\n======= Personal Finance Tracker =======")
        print("1. Add Transactions")
        print("2. View All Transactions")
        print("3. Search Transactions")
        print("4. View Financial Summary")
        print("5. View Monthly Spending Chart")
        print("6. Delete data")
        print("7. Save & Exit")
        print("========================================")

        choice = input("Enter your choice (1-7): ")

        if choice == '1':
            add_transaction()
        elif choice == '2':
            view_transactions()
        elif choice == '3':
            search()
        elif choice == '4':
            finance_summary()
        elif choice == '5':
            display_monthly_chart(transactions)
        elif choice=='6':
            delete_data()
        elif choice == '7':
            print("Goodbye! 👋")
            save_data()
            break # This break exits the main while loop
        else:
            print("\n❌ Invalid choice. Please select an option from 1 to 7.")

# The main() function is called only once here
if __name__ == "__main__":
    main()
```
